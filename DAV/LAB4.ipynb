```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('marketing_and_product_performance.csv')
```

```python
print(data.head())
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Expenses', y='Sales', data=data)

plt.title('Scatter Plot of Marketing Expenses vs Sales')
plt.xlabel('Marketing Expenses')
plt.ylabel('Sales')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(x='Marketing_Expenses', y='Sales', data=data, scatter=False, color='red')

plt.title('Regression Line for Marketing Expenses vs Sales')
plt.xlabel('Marketing Expenses')
plt.ylabel('Sales')

plt.show()
```
