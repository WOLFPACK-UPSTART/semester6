```python
# Scatter Plots – Bivariate Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
customer_data = pd.read_csv('customer_data.csv')

# Scatter plot between Age and Income
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Income', data=customer_data)
plt.title('Scatter Plot between Age and Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# Scatter plot between Spending_Score and Purchase_Frequency
plt.figure(figsize=(8,6))
sns.scatterplot(x='Spending_Score', y='Purchase_Frequency', data=customer_data)
plt.title('Scatter Plot between Spending Score and Purchase Frequency')
plt.xlabel('Spending Score')
plt.ylabel('Purchase Frequency')
plt.show()
```

```python
# Inferential Statistics
import scipy.stats as stats

# Example: t-test to compare means of Income for two groups based on Membership_Years
median_membership = customer_data['Membership_Years'].median()
group1 = customer_data[customer_data['Membership_Years'] <= median_membership]['Income']
group2 = customer_data[customer_data['Membership_Years'] > median_membership]['Income']

# Perform independent t-test
result = stats.ttest_ind(group1, group2)
print('T-test result:', result)
```

```python
# Descriptive Statistics
# Summary statistics
summary_stats = customer_data.describe()
print(summary_stats)
```

```python
# Dashboard Evaluation (using Streamlit)
# Save this cell as app.py and run with: streamlit run app.py

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('customer_data.csv')

st.title('Customer Data Dashboard')

# Show data summary
st.write('Data Summary:')
st.write(data.describe())

# Scatter plot
st.write('Scatter Plot: Age vs Income')
fig, ax = plt.subplots()
sns.scatterplot(x='Age', y='Income', data=data, ax=ax)
st.pyplot(fig)

# Histogram
st.write('Histogram of Spending Score')
fig2, ax2 = plt.subplots()
data['Spending_Score'].hist(ax=ax2)
st.pyplot(fig2)
```

```python
# Univariate Analysis - Histogram for Age
plt.figure(figsize=(8,6))
plt.hist(customer_data['Age'], bins=30, color='blue', alpha=0.7)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Univariate Analysis - Boxplot for Income
plt.figure(figsize=(8,6))
plt.boxplot(customer_data['Income'])
plt.title('Boxplot of Income')
plt.ylabel('Income')
plt.show()
```

```python
# Correlation Analysis
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
corr_matrix = customer_data.corr()

# Heatmap of correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Feature Engineering
# Example: Creating a new feature - Average Purchase Amount
customer_data['Average_Purchase_Amount'] = customer_data['Last_Purchase_Amount'] / customer_data['Purchase_Frequency']

# Display first few rows
print(customer_data.head())
```

```python
# Data Cleaning

# Check for missing values
missing_values = customer_data.isnull().sum()
print('Missing values per column:')
print(missing_values)

# Fill missing values with median for numerical columns
customer_data.fillna(customer_data.median(), inplace=True)

# Remove duplicates
customer_data.drop_duplicates(inplace=True)

# Display cleaned data info
print(customer_data.info())
```
