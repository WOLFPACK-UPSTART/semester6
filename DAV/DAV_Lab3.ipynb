```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset from Google Drive
file_path = '/content/drive/MyDrive/Colab Notebooks/performance.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
data.head()
```

```python
# Set the style of the plots
sns.set(style="whitegrid")
```

```python
# Histogram with a KDE overlay for exam scores
plt.figure(figsize=(12, 6))
sns.histplot(data['EXAM SCORE'], bins=30, kde=True, color='skyblue', edgecolor='black')

# Titles and labels
plt.title('Distribution of Exam Scores with KDE Overlay', fontsize=16)
plt.xlabel('Exam Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Showing the plot
plt.show()
```
