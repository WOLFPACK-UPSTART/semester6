```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "sales_data.csv"
data = pd.read_csv(file_path)
```

```python
# Central Tendency Analysis
print("Central Tendency:")
print("Mean:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].mean())
print("\nMedian:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].median())
print("\nMode:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].mode())

# Central Tendency - Bar Plot of Mean
means = data[["Price Each", "Cost price", "turnover", "margin"]].mean()
means.plot(kind='bar', figsize=(10, 6), title="Mean Values for Numerical Features")
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Variability Analysis
print("Variability Analysis:")
print("Standard Deviation:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].std())
print("\nVariance:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].var())

# Variability - Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[["Price Each", "Cost price", "turnover", "margin"]])
plt.title("Box Plot of Numerical Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Distribution Analysis
print("Distribution Analysis:")
print(data[["Price Each", "Cost price", "turnover", "margin"]].describe())

# Distribution - Histograms for Numerical Features
data[["Price Each", "Cost price", "turnover", "margin"]].hist(bins=20, figsize=(12, 8), grid=False, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```
